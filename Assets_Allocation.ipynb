## Load log data from google spreadsheet

The "Log" worksheet of the "Asset Allocation" spreadsheet is loaded to data frame.

```python
## Read data from 'Log' tab
LOG = SPREADSHEET.worksheet('Log').get_all_values()[1:]
LOG = pd.DataFrame.from_records(LOG, columns=["date", "ticker", "type", "price", "shares", "fee", "note", "gain"])
LOG.date = pd.to_datetime(LOG.date, format="%b %d, %Y")

def toFloat(x):
    return 0 if x == "" else float(x.replace("$","").replace(",",""))

def toInt(x):
    return 0 if x == "" else int(x)

LOG.price  = LOG.price.apply(toFloat)
LOG.fee    = LOG.fee.apply(toFloat)
LOG.shares = LOG.shares.apply(toInt)
LOG.tail()
```

## Simulate with Log

```python
## Download data from Yahoo Finance
tickers = list(LOG.ticker.unique())
DATA = pdr.DataReader(tickers, data_source="yahoo", start=LOG.date[0])["Adj Close"] # VOO
```

```python
SIM = {}
```

```python
assets = {}
prevMonth = None
cash = 0
gainAndLoss = 0.0

for ticker in tickers:
    assets[ticker] = {"avgCost": 0.0, "avgFee": 0.0, "shares": 0}

for i, trade in LOG.iterrows():
    thisMonth = trade.date.strftime("%Y-%m")
    if thisMonth not in SIM:
        SIM[thisMonth] = {"div": 0.0, "fee": 0.0, "gain": 0.0, "assets": None, "cash": 0.0}
        if prevMonth is not None:
            SIM[prevMonth]["assets"] = copy.deepcopy(assets)
            SIM[prevMonth]["cash"] = cash
        prevMonth = thisMonth

    if trade.type == "Deposit":
        print("Deposit", trade.price)
        cash += trade.price

    elif trade.type == "Withdraw":
        print("Withdraw", trade.price)
        cash -= trade.price

    if trade.type == "Buy":
        avgCost = assets[trade.ticker]["avgCost"]
        avgFee  = assets[trade.ticker]["avgFee"]
        shares  = assets[trade.ticker]["shares"]
        avgCost = (avgCost * shares + trade.price * trade.shares) / (shares + trade.shares)
        assets[trade.ticker]["avgCost"] = avgCost
        assets[trade.ticker]["shares"]  = shares + trade.shares
        assets[trade.ticker]["avgFee"]  = (avgFee * shares + trade.fee) / (shares + trade.shares)
        assert(assets[trade.ticker]["shares"] > 0)

        cash -= (trade.price * trade.shares + trade.fee)

    elif trade.type == "Sell":
        avgCost = assets[trade.ticker]["avgCost"]
        avgFee  = assets[trade.ticker]["avgFee"]
        assets[trade.ticker]["shares"] -= trade.shares
        assert(assets[trade.ticker]["shares"] >= 0)

        if assets[trade.ticker]["shares"] == 0:
            assets[trade.ticker]["avgCost"] = 0
            assets[trade.ticker]["avgFee"]  = 0
            assets[trade.ticker]["shares"]  = 0

        gain = (trade.price - (avgCost + avgFee)) * trade.shares - trade.fee
        #print(trade.date, trade.ticker, gain, SIM[thisMonth]["gain"])
        gainAndLoss += gain
        SIM[thisMonth]["gain"] += gain

        cash += (trade.price * trade.shares - trade.fee)
    
    elif trade.type == "Div":
        SIM[thisMonth]["div"] += (trade.price - trade.fee)
        cash += (trade.price - trade.fee)

    if trade.type in ["Buy", "Sell"]:
        SIM[thisMonth]["fee"] += trade.fee

SIM[thisMonth]["assets"] = copy.deepcopy(assets)
SIM[thisMonth]["cash"]   = cash
SIM[thisMonth]["cash"]
```

```python
#
# Benchmark
#
BENCH = {}
prevMonth = None
iter = []
for i, trade in LOG.iterrows():
    thisMonth = trade.date.strftime("%Y-%m")
    if thisMonth not in BENCH:
        BENCH[thisMonth] = {"shares": 0, "deposit": 0.0}
        iter.append(thisMonth)
        if prevMonth is not None:
            BENCH[thisMonth]["shares"]  = BENCH[prevMonth]["shares"]
            BENCH[thisMonth]["deposit"] = BENCH[prevMonth]["deposit"]
        prevMonth = thisMonth

    if trade.type == "Deposit":
        BENCH[thisMonth]["shares"]  += trade.price / DATA["VOO"][trade.date]
        BENCH[thisMonth]["deposit"] += trade.price

    elif trade.type == "Withdraw":
        BENCH[thisMonth]["shares"]  -= trade.price / DATA["VOO"][trade.date]
        BENCH[thisMonth]["deposit"] -= trade.price

list_benchmark = []
for it in iter:
    list_benchmark.append(DATA["VOO"][it][-1] * BENCH[it]["shares"])

#
# Mine
#
list_cash = []
list_assets = []
list_total = []
for thisMonth in iter:
    #thisMonth = it.strftime("%Y-%m")
    cash = SIM[thisMonth]["cash"]
    assets = SIM[thisMonth]["assets"]

    totalAssets = 0
    for ticker in assets.keys():
        if ticker == "Cash":
            continue
        if assets[ticker]["shares"] == 0:
            continue
        totalAssets += DATA[ticker][thisMonth][-1] * assets[ticker]["shares"]
    list_cash.append(cash)
    list_assets.append(totalAssets)
    list_total.append(cash + totalAssets)
    #print(thisMonth, totalAssets, cash)

#
# Rate of Returns
#
rr_benchmark = []
rr_mine = []
for i, thisMonth in enumerate(iter):
    rr_benchmark.append(list_benchmark[i] / BENCH[thisMonth]["deposit"] - 1)
    rr_mine.append(list_total[i] / BENCH[thisMonth]["deposit"] - 1)

#
# Draw chart
#
index = pd.period_range(start=iter[0], end=TODAY, freq="M")
fig = pd.DataFrame({"benchmark": list_benchmark, "mine": list_total}, index=index)
fig.index = fig.index.strftime('%b %d, %Y')
fig[-12:].plot_bokeh.line(figsize=(1500,300), legend="top_left", rangetool=False, panning=False, zooming=False, ylabel="USD ($)", title="assets", colormap=["#ea4335", "#4285f4"])

fig3 = pd.DataFrame({"stock": list_assets, "cash": list_cash}, index=index)
fig3.index = fig3.index.strftime('%b %d, %Y')
fig3[-12:].plot_bokeh.bar(figsize=(1500,300), legend="top_left", rangetool=False, panning=False, zooming=False, stacked=True, ylabel="USD ($)", colormap=["#4285f4", "#b7b7b7"])
```

```python
fig2 = pd.DataFrame({"benchmark": rr_benchmark, "mine": rr_mine}, index=index)
fig2.index = fig2.index.strftime('%b %d, %Y')
fig2.apply(lambda x: x*100)[-12:].plot_bokeh.bar(figsize=(1500,300), legend="top_left", rangetool=False, panning=False, zooming=False, ylabel="Rate of Return (%)", colormap=["#ea4335", "#4285f4"])
fig2.pct_change().apply(lambda x: x*100)[-12:].plot_bokeh.bar(figsize=(1500,300), legend="top_left", rangetool=False, panning=False, zooming=False, ylabel="MoM (%)", colormap=["#ea4335", "#4285f4"])
```

```python
def get_mdd(x):
    arr_v = np.array(x)
    arr_max = np.maximum.accumulate(arr_v)
    peak_lower = np.argmax(arr_max - arr_v)
    peak_upper = np.argmax(arr_v[:peak_lower])
    return (x[-1] - arr_max[-1]) / arr_max[-1], (arr_v[peak_lower] - arr_v[peak_upper]) / arr_v[peak_upper]

mdds = []
nows = []
for ticker in tickers:
    now, mdd = get_mdd(data[ticker])
    mdds.append(mdd * 100)
    nows.append(now * 100)
    #print("{}\t{:8.2f} {:8.2f}".format(ticker, now * 100, mdd * 100))
fig = pd.DataFrame({"MDD (2021)": mdds, "Now": nows}, index = tickers)
fig.plot_bokeh.bar(figsize=(1500,300), legend="bottom_left", rangetool=False, panning=False, zooming=False)
```
